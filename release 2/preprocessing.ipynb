```python
df_clean = df.copy()

print("\n" + "="*50)
print("–ò–ù–î–ò–í–ò–î–£–ê–õ–¨–ù–ê–Ø –û–ë–†–ê–ë–û–¢–ö–ê –ö–û–õ–û–ù–û–ö:")
print("="*50)

print(f"\n1. Unnamed: 0:")
print(f"   üóëÔ∏è  –£–¥–∞–ª—è–µ–º —Ç–µ—Ö–Ω–∏—á–µ—Å–∫—É—é –∫–æ–ª–æ–Ω–∫—É")
df_clean = df_clean.drop('Unnamed: 0', axis=1)
```

```python
print(f"\n2. Number_of_rooms:")
if 'Number_of_rooms' in df_clean.columns:
    missing_count = df_clean['Number_of_rooms'].isna().sum()
    print(f"   –ü—Ä–æ–ø—É—Å–∫–∏: {missing_count}")
    if missing_count > 0:
        # –°–Ω–∞—á–∞–ª–∞ –∏–∑–≤–ª–µ–∫–∞–µ–º —á–∏—Å–ª–æ–≤–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ –∏–∑ —Å—Ç—Ä–æ–∫–∏
        df_clean['rooms_clean'] = df_clean['Number_of_rooms'].astype(str).str.extract('(\d+)').astype(float)
        mode_rooms = df_clean['rooms_clean'].mode()[0]
        df_clean['rooms_clean'] = df_clean['rooms_clean'].fillna(mode_rooms)
        print(f"   –ó–∞–ø–æ–ª–Ω–µ–Ω–æ {missing_count} –ø—Ä–æ–ø—É—Å–∫–æ–≤ –º–æ–¥–æ–π: {mode_rooms}")
else:
    print(f"–ö–æ–ª–æ–Ω–∫–∞ –Ω–µ –Ω–∞–π–¥–µ–Ω–∞")
```

```python
print(f"\n12. Renovation:")
if 'Renovation' in df_clean.columns:
    missing_count = df_clean['Renovation'].isna().sum()
    print(f"   –ü—Ä–æ–ø—É—Å–∫–∏: {missing_count}")
    if missing_count > 0:
        mode_renovation = df_clean['Renovation'].mode()[0] if not df_clean['Renovation'].mode().empty else 'Not specified'
        df_clean['Renovation'] = df_clean['Renovation'].fillna(mode_renovation)
        print(f"    –ó–∞–ø–æ–ª–Ω–µ–Ω–æ {missing_count} –ø—Ä–æ–ø—É—Å–∫–æ–≤ –º–æ–¥–æ–π: '{mode_renovation}'")
else:
    print(f"    –ö–æ–ª–æ–Ω–∫–∞ –Ω–µ –Ω–∞–π–¥–µ–Ω–∞")
```

```python
print(f"\n13. Room_area:")
if 'Room_area' in df_clean.columns:
    missing_count = df_clean['Room_area'].isna().sum()
    print(f"   –ü—Ä–æ–ø—É—Å–∫–∏: {missing_count}")
    if missing_count > 0:
        df_clean['Room_area'] = df_clean['Room_area'].fillna('Not specified')
        print(f"   –ó–∞–ø–æ–ª–Ω–µ–Ω–æ {missing_count} –ø—Ä–æ–ø—É—Å–∫–æ–≤ –∑–Ω–∞—á–µ–Ω–∏–µ–º: 'Not specified'")
else:
    print(f"   –ö–æ–ª–æ–Ω–∫–∞ –Ω–µ –Ω–∞–π–¥–µ–Ω–∞")
```

```python
print(f"\n14. Balcony:")
if 'Balcony' in df_clean.columns:
    missing_count = df_clean['Balcony'].isna().sum()
    print(f"   –ü—Ä–æ–ø—É—Å–∫–∏: {missing_count}")
    if missing_count > 0:
        df_clean['Balcony'] = df_clean['Balcony'].fillna('No')
        print(f"   –ó–∞–ø–æ–ª–Ω–µ–Ω–æ {missing_count} –ø—Ä–æ–ø—É—Å–∫–æ–≤ –∑–Ω–∞—á–µ–Ω–∏–µ–º: 'No'")
else:
    print(f"   –ö–æ–ª–æ–Ω–∫–∞ –Ω–µ –Ω–∞–π–¥–µ–Ω–∞")
```
